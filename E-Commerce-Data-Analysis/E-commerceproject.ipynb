```python
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio 
import plotly.colors as colors 
pio .templates.deafault = 'plotly_white'
```

```python
data = pd.read_csv("Sample - Superstore.csv", encoding = 'latin-1') #latine usebecause handle special characters 
```

# converting date columns 

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week' ] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

 # Monthly sales analysis 

```python

sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month ,
              x='Order Month',
              y='Sales',
              title='Monthly Sales Analysis')
```

```python
import matplotlib.pyplot as plt

# Group data by month and calculate total sales
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()

# Create a line plot
plt.plot(sales_by_month['Order Month'], sales_by_month['Sales'], marker='o')

# Add title and labels
plt.title('Monthly Sales Analysis')
plt.xlabel('Order Month')
plt.ylabel('Sales')

# Display the plot
plt.show()
```

```python
sales_by_month
```

# Sales by Category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
categories = sales_by_category['Category']
sales = sales_by_category['Sales']

plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    sales,
    labels=categories,
    autopct='%1.1f%%',  # Show percentage inside the pie
    startangle=90,      # Rotate the pie chart for better readability
    colors=plt.cm.Pastel1.colors,  # Use a pastel color scheme
    wedgeprops={'width': 0.5},  # Create a donut chart by setting the hole size
    textprops={'fontsize': 12, 'color': 'black'}  # Customize text properties
)

plt.title('Sales Analysis by Category', fontsize=24)

plt.axis('equal')

plt.show()
```

```python
profit_by_category
```

```python
# Assuming sales_by_category is a DataFrame with 'Category' and 'Profit' columns
categories = profit_by_category['Category']
profits = profit_by_category['Profit']

# Create a pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    profits,
    labels=categories,
    autopct='%1.1f%%',  # Show percentage inside the pie
    startangle=90,      # Rotate the pie chart for better readability
    colors=plt.cm.Pastel1.colors,  # Use a pastel color scheme
    wedgeprops={'width': 0.5},  # Create a donut chart by setting the hole size
    textprops={'fontsize': 12, 'color': 'black'}  # Customize text properties
)

# Add a title
plt.title('Profit Analysis by Category', fontsize=24)

# Equal aspect ratio ensures that the pie is drawn as a circle
plt.axis('equal')

# Display the plot
plt.show()
```

# Sales and Profit - customer segment 

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

plt.figure(figsize=(10, 6))
plt.bar(sales_profit_by_segment['Segment'], sales_profit_by_segment['Sales'], color='skyblue', label='Sales')
plt.bar(sales_profit_by_segment['Segment'], sales_profit_by_segment['Profit'], color='lightgreen', label='Profit', alpha=0.7)

plt.title('Sales and Profit Analysis by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Amount')
plt.legend()

plt.show()
```

# Sales to profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales_to_Profit_Ratio']])
```
